# **Linked List in Python**



Linked list is a linear data structure in which the elements are stored in contigeous memory location or they are scattered anywhere in the memory.

Each element of a linked list is called a **node**, and every node has two different fields:


1.   **Data** contains the value to be stored in the node.
2.   **Next** contains a reference to the next node on the list.


Next contains a reference to the next node on the list.

The **first node** is called the **head**, and itâ€™s used as the starting point for any iteration through the list.

 The **last node** must have its next reference pointing to **None** to determine the end of the list.

```python
#Create a class of Node 
class Node:
  def __init__(self,data=None,next=None):
    self.data = data
    self.next = next

#Create a LinkedList
class LinkedList :
  def __init__(self):
    self.head = None

#Create a method to print a LinkedList
  def show(self):
    node = self.head
    llstr = ''
    while node is not None :
      llstr += str(node.data) + '-->'
      node = node.next
    return llstr

 #Create a Method to add  Elements at the End    
  def add_End(self,new):
    new_node = Node(new)
    node = self.head
    
    if self.head is None :
      self.head = new_node

    while node.next is not None :
      node = node.next
    
    node.next = new_node

    # itr = self.head 
    # while itr.next :
    #   itr = itr.next
    # itr.next = new_node

#Create a Method to add Elements at the begining   
  def add_First(self,new):
    new_node = Node(new)
    new_node.next = self.head
    self.head = new_node

#Create a Method to count lenth of Nodes in LinkedList
  def Count_Node(self):
    if self.head is None :
      raise Exception ('No Node  Found ..!! List is Empty')
    node = self.head
    count = 0
    while node :
      count += 1
      node = node.next
  
    return count

#Create a Method to check whether LinkedList is empty or not.
  def ll_isEmpty(self):
    if self.head is None :
      print("Linked List is Empty")

#Create a Method to check value at  the given index in LinkedList.
  def value_at(self, ind):
    
    if(ind<0 or ind>=self.Count_Node()):
      raise Exception('Range is Out of index ')

    else:
      count = 0
      node = self.head
      while count < ind :
         node = node.next
         count += 1
      return node.data    

#Create a Method to get Value from First Node
  def value_at_front(self):
    node = self.head
    if node is None:
       raise Exception('No Node Found !! List is Empty')
    
    return node.data

#Create a Method to get Value at Last Node
  def value_at_end(self):
    node = self.head
    if node is None:
      raise Exception('No Node Found !! List is Empty')


    while node.next:
      node = node.next
    return node.data     

#Create a Method to pop the Element at the Begining 
  def pop_front(self):
    if self.head is None:
       raise Exception('No Node Found !! List is Empty')

    
    node = self.head
    self.head = self.head.next
    node.next = None
    return node.data + ' removed..!!'
    
#Create a Method to pop the Element at the End
  def pop_end(self):
    if self.head is None:
      raise Exception('No Node Found..!! List is Empty')

    
    node = self.head
    count = 1
    while count < self.Count_Node() -1  :
      count += 1
      node = node.next
    
    elePop = node.next.data
    node.next = None

    return  elePop + ' removed ..!!'

#Create a Method to insert the node at given index
  def insert_at (self,ind,data):
    new_node = Node(data)

    if self.head is None :
      raise Exception('LinkedList is Empty')
    elif ind > self.Count_Node() or ind < 0:
      raise Exception('Out of range Index ')
    elif ind == 0:
      self.head = new_node
      return

    count = 0
    node = self.head
    while node:
      if count == ind - 1:
        new_node.next = node.next
        node.next = new_node   
        break
      node = node.next  
      count  += 1

#Create a Method to remove the node at given index
  def remove_at(self,ind):

    if self.head is None :
      raise Exception('LinkedList is Empty..!!')
    elif ind > self.Count_Node() or ind < 0:
      raise Exception('Out of range Index..!!')

    count = 0
    node = self.head
    while node:
      if count == ind - 1:
        temp_node = node.next
        self.head.next = temp_node.next
        break
      node = node.next  
      count  += 1
      
#Create a Method to return the value  of the node at nth position from the end of the list
  def count_from(self,ind):
    if self.head is None:
      raise Exception('LinkedList is Empty..!!')
    if ind >= self.Count_Node() or ind < 0 :
      raise Exception ('Out of range Index..!!')
    
    count = 0
    node = self.head

    while node :
      if count == ind:
        temp = 0
        while node :
          node = node.next
          temp += 1
        return temp
      node = node.next
      count += 1

#Create a Method to Convert list in to reversal order
  def reverse_list(self):
    if self.head is None:
      raise Exception('No List Found ..!!')

    node = self.head
    prev = None
    while node:
      next = node.next
      node.next = prev
      prev = node
      node = next
    self.head = prev

  #Create a method to  removes the first item in the list with this value
  def remove_value(self,data):
    if self.head is None:
      raise Exception('No list Found..!!')
    
    node = self.head
    prev = None
    while node:
      if (node.data == data):
        temp = node.next
        node.next = None
        prev.next = temp
        return
      prev = node
      node = node.next


```
